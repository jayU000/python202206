```python
stars = soup.select('li.quickResultLstCon')

import pandas as pd

star_data = []

for star in stars :
  # 
  name = star['data-name']
  # 위도
  lat = star['data-lat']    # 키값 밸류값 형태의 딕셔너리 추출
  # 경도
  lng = star['data-long']
  # 주소
  addr = str(star.select('p.result_details')[0]).split('<br/>')[0].split('>')[1]
  # 전화번호
  tel = str(star.select('p.result_details')[0]).split('<br/>')[1].split('<')[0]

  star_data.append([name, lat, lng, addr, tel])

df = pd.DataFrame(star_data, columns=['지점명','위도', '경도', '주소', '전화번호'])

df2 = df.drop_duplicates()
print(df2)
```

```python
import pandas as pd
```

```python
# 서울시 구별 스타벅스 매장수 파악

sgg_names = []

for addr in df2['주소'] :
  sgg =addr.split()[1]
  sgg_names.append(sgg)

df2['시군구명'] = sgg_names
print(df2)

freq = df2.groupby(['시군구']).count() 
print(freq)
```

```python
# 좀더 보기 좋게 표현

star_count = df2.pivot_table(index='시군구명', values='지점명', aggfunc='count').rename(columns={'지점명':'매장수'})

star_count
```

```python
# 서울시 구별 인구수, 종사자 수 읽어오기

seoul_pop = pd.read_excel('sgg_pop.xlsx', thousands=',')
seoul_biz = pd.read_excel('sgg_biz.xlsx')

seoul_pop
```

```python
# 세 개의 데이터 프레임을 하나로 병합

seoul_sgg = pd.merge(seoul_pop, seoul_biz, how = 'left', on = '시군구명')
seoul_sgg

seoul_sgg = pd.merge(seoul_sgg, star_count, how='right',on='시군구명')
seoul_sgg
```

```python
# 주민등록 인구 십만명당 매장 수

seoul_sgg['거주자_십만명당_매장수'] = seoul_sgg['매장수'] / (seoul_sgg['주민등록인구']/100000)
seoul_sgg
```

```python
# 종사자수 십만명당 매장 수

seoul_sgg['종사자수_십만명당_매장수'] = seoul_sgg['매장수'] / (seoul_sgg['종사자수']/100000)
seoul_sgg
```

```python
# 시업체수 만개당 매장 수

seoul_sgg['시업체수_만개당_매장수'] = seoul_sgg['매장수'] / (seoul_sgg['사업체수']/10000)
seoul_sgg
```
