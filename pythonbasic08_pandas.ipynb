```python
# 시리즈와 데이터프레임에 대해 자세히 알아보자

# 시리즈 만들기

import pandas as pd

s = pd.Series(['apple','banana','cherry'])
print(type(s))
print(s)

s2 = pd.Series(['apple','banana','cherry'], index=['a','b','c'])
print(s2)


# 데이터 프레임 만들기

person = pd.DataFrame({
    'name':['Austin', 'Kelly'],
    'job':['prof','chemist'],
    'born':['1910-07-25','1890-06-13'],
    'age':[31,52]
})

print(person)
print(type(person))

# 시리즈 처럼 데이터프레임도 인덱스를 따로 지정하지 않으면 
# 인덱스를 0부터 자동으로 생성한다. 


```

```python
# 인덱스 값 직접 지정

person = pd.DataFrame({
    'family':[5,4],
    'job':['prof','chemist'],
    'name':['Austin', 'Kelly'],
    'job':['prof','chemist'],
    'born':['1910-07-25','1890-06-13'],
    'age':[31,52]}, index=['first', 'second'])

print(person)
print(type(person))

# 딕셔너리는 순서를 보장하지 않는다. 
```

```python
from collections import OrderedDict
# 따라서 순서까지 보장된 딕려너리 전달하려면...OrderedDict 모듈을 사용한다. 
# 딕셔너리의 데이터 순서를 그대로 유지하면서 데이터프레임을 만들수 있다. 

person = pd.DataFrame(OrderedDict({
    'family':[5,4],
    'job':['prof','chemist'],
    'name':['Austin', 'Kelly'],
    'job':['prof','chemist'],
    'born':['1910-07-25','1890-06-13'],
    'age':[31,52]
}))


print(person)
print(type(person))
```

```python
# 판다스의 데이터 구성의 가장 기본은 시리즈이다. 

# 데이터 프레임에서 시리즈를 선택 선택하려면 loc 속성에 인덱스를 전달한다. 

person = pd.DataFrame({
    'family':[5,4],
    'job':['prof','chemist'],
    'name':['Austin', 'Kelly'],
    'job':['prof','chemist'],
    'born':['1910-07-25','1890-06-13'],
    'age':[31,52]}, index=['first', 'second'])

first_one = person.loc['first'] # loc 는 인덱스 값으로 선택한다. 
type(first_one)
```

```python
sci = pd.read_csv('scientists.csv')
sci.head()
```

```python
ages = sci['Age']
ages.max()
```

```python
ages.mean()
```

```python
# 조건 추출

ages[ages > ages.mean()]
```

```python
# ages > ages.mean() 의 결과값의 개수가 여러개이다. 이렇게 시리즈나 데이터프레임에 있는 모든 데이터에
# 대해 한번에 연산하는 것을 broadcasting 브로드캐스팅이라고 한다. 
# 시리즈 처럼 여러개 값을 가진 데이터를 벡터 vector 라고 하고 크기를 나타내는 데이터는 스칼라 scaler라고 부른다. 
```

```python
ages + ages  

# 같은 길이의 벡터로 더하기 연산을 수행하였다. 결과로 벡터가 출력된다. 
```

```python
ages * ages

# 같은 길이의 벡터로 곱하기 연산을 수행하였다. 결과로 벡터가 출력된다. 
```

```python
ages + 100 # 벡터에 스칼라를 더해 보자
```

```python
ages * 2 # 벡터에 스칼라를 곱해 보자
```

```python
# 길이가 다른 벡터를 연산하면 어떻게 될까...
# 시리즈와 시리즈의 연산은 같은 인덱스의 값만 계산한다. 

ages + pd.Series([1,100])   # 2 개와 7개 로 길이가 다를 경우 인덱스가 일치한 값들만 계산된다. 

# NaN  은 누락값 / 결측치 로 처리된다. 
```

```python
sci[sci['Age'] > sci['Age'].mean()]
```

```python
# 데이터 프레임에서 스칼라 연산

sci * 2

# sci데이터 프레임에 2를 곱하면 정수는 2를 곱한 숫자가 되고 
# 문자열은 2번 반복 출력된다. 
```
